## apply() :  allows to apply function to every value in the column

```python
df['Platform'].unique()
```

```python
sony = ["PS","PS2","PS3","PS4","PSP","PSV"]
nintendo = ["N64","GB","GBA","GC","Wii","WiiU","DS","NES","3DS","SNES"]
microsoft = ["XB","X360","Xone"]
pc = ["PC"]
atari = ["2600"]
panasonic = ["3D0"]
sega = ["DC","GC","SAT","SCD"]
```

```python
def categories(x):
    """
    This function maps the products to the platform makers and will be coupled with .apply
    
    """
    if x in sony:
        return "Sony"
    elif x in nintendo:
        return "Nintendo"
    elif x in microsoft:
        return "Microsoft"
    elif x in pc:
        return "PC"
    elif x in atari:
        return "Atari"
    elif x in panasonic:
        return "Panasonic"
    elif x in sega:
        return "Sega"
    else:
        return "Other"
```

```python
# now we can take a column and apply the categories function to the same

df['Platform'].apply(categories).value_counts().reset_index()
```

```python
# we can create a new column called Platform Makers 

df['Platform_Makers'] = df['Platform'].apply(categories)
df.head()
```

## Merging 


```python
# Inner join
```

```python
import pandas as pd
orders = pd.read_excel("Sales Order.xlsx",sheet_name='Orders')
returns = pd.read_excel("Sales Order.xlsx",sheet_name='Returns')
```

```python
print(orders.shape)
print(returns.shape)
print(orders.head())
print(returns.head())
```

```python
# apply inner join based on Order ID
# by default pd.merge takes inner join 

inner_join = pd.merge(left=orders,
                     right=returns,
                     left_on='Order ID', # on ="Order ID"
                     right_on = 'Order ID',
                     how='inner')
inner_join.head()
```

```python
# apply inner join based on Order ID
# by default pd.merge takes inner join 

inner_join_new = pd.merge(left=orders,
                     right=returns,
                      on="Order ID")
inner_join_new.head()
```

```python
inner_join_new.shape
```

```python
inner_join.shape
```

```python
# _x means left dataframe
# _y means right dataframe
```

```python
inner_join = pd.merge(left=orders,
                      right=returns,
                      left_on='Order ID',
                      right_on = 'Order ID',
                      how='inner',
                      suffixes=("_Orders","_Returns")
                     )
inner_join.head()
```

```python
# easier way to find matching records in both the tables

orders['Order ID'].isin(returns['Order ID']).value_counts()

# the output shows that there are 89 matching records
```

```python
returns.columns
```

```python
returns.rename(columns={returns.columns[0]:'Invoice ID'},inplace=True )
returns.head()

# rename order ID to invoice ID
```

```python
# perform inner join

inner_join_new = pd.merge(orders,returns,left_on='Order ID',right_on='Invoice ID',how='inner')
inner_join_new.head()
```

```python
inner_join_new.shape
```

```python
# now, we have 12 columns as opposed to erlier inner join output shape which had 11 columns

# because we renamed the order ID to Invoice ID, hence it is treating it as other column

# now let us see how to perform join on 2 primary keys,i.e, Order ID and Invoice ID (data is same in both columns)
```

```python
inner_join_again = pd.merge(orders,returns,
                            left_on=['Order ID','City'],
                           right_on=['Invoice ID','City'],
                           )
```

```python
inner_join_again.shape
```

```python
inner_join_again = pd.merge(orders,returns,
                            left_on=['Order ID','City'],
                           right_on=['Invoice ID','City'],
                            indicator=True
                           )

# indicator = True , tells us the status of records as to which dataset are they originating from
# either from left or from right
# _merge column gets appended 

inner_join_again.head()
```


## Left Join

```python
### All records from the left dataset and common records from the right dataset
```

```python
returns.rename(columns={'Invoice ID' : 'Order ID'}, inplace=True)

returns.head()
```

```python
left_join = pd.merge(left=orders,
                    right=returns,
                    on='Order ID',
                    how='left',
                    indicator=True)
left_join.head()
```

```python
left_join.shape
```

```python
orders.shape
```

```python
# the shape/ number of records from orders and left join will remain the same
# because left join will all the records from the left table, in this case which is orders
# and matching ones from the right
```

```python
left_join._merge.value_counts()
```

## Right Join

```python
# all the records from right table and matching records from the left table
#  right only + inner join
```

```python
right_join = pd.merge(orders,returns,on='Order ID',how='right',indicator=True)

right_join.head()
```

```python
right_join._merge.value_counts()

# right_only + join
```

```python
returns.shape
```

```python
right_join.shape
```

## Full Join

```python
### All the records from the left dataset/table + All the records from the right dataset/Table - common records in both tables
```

```python
# (A + B) - (A intersection B)
```

```python
full_join = pd.merge(orders,returns,on='Order ID',how='outer',indicator=True)
print(full_join.head())
print(full_join.shape)
```

```python
print(orders.shape, returns.shape)
```

```python
full_join._merge.value_counts()
```

```python
7000 - 89 #89 are common records
```

```python
full_join[full_join._merge.isin(['right_only','both'])].shape # RIGHT JOIN
```
