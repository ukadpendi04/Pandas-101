### Return the first 10 employees hired from March 1, 2010

```python
# use kind='mergesort' if sorting
```

```python
def first_tenemployees(df):
    """This function returns the first 10 employees hired from March 1, 2010"""
    
    return (df.query('hire_date >= "3-1-2010"')
            .sort_values('hire_date',kind='mergesort')
            .head(10))

#kind = mergesort, is a stable sort
#if two employees have a same hire date they will be displayed in the same order
# it will be stable sorting
```

```python
first_tenemployees(emp)
```

```python
def top_employees(df):
    """
    This functions returns the top employees of each dept --> by salary
    
    Returns
    -------
    DataFrame : Department, title, salary
    """
    
    cols = ['dept','title','salary']
    
    return df[cols].sort_values('salary',ascending=False,kind='mergesort')
```

### Return the Median salary for all employees hired from Jan 1,2010

```python
def median_salaryJan(df):
    """This function returns the median salary of employees hired from Jan 1,2010"""
    
    filterCondition = df['hire_date'] >= '2010-1-1'
    
    return df.loc[filterCondition,'salary'].median()

# .loc goes will go as per the filter, and then apply median on salary column
```

### Return the number of employees from Police department, that do not have the word "police" in their title, ignore case

```python
def police_employees(df):
    """This functions returns employees who do not have police their title and work for police dept"""
    
    filter_police = df['dept'] == "Police"
    
    filter_title = ~df['title'].str.contains('police',case=False)
    # this will return true if the title contains police, hence we insert a ~ in the beginnning which will reverse the case
    
    final_filter = filter_police & filter_title #intersection of two filters, checking if both will be true
    
    return final_filter.sum()
```

```python
# from police but do not have word police in their title

police_employees(emp)
```

### Return a dataframe in strictly increasing salaries, without changing the row order, but filtering such that the employee who does not have salary greater than employee above them

```python
def salary_employees(df):
    """This function returns the salary hierarchy amongst employees"""
    
    # we will drop the missing values in salary column
    
    df = df.dropna(subset=['salary'])
    # in order to look at maximum salary increasing over time, we can use cummax() - cummulative max function
    
    sal = df['salary']
    return sal.cummax()
    
```

```python
# now we will only keep the salaries, which are greater than the salary above it
# look at each salary and test that it is greater than cummax()
# only then keep that

# we can try using shift method which gives us an understanding in depth

def salary_employees(df):
    """This function returns the salary hierarchy amongst employees"""
    
    # we will drop the missing values in salary column
    
    df = df.dropna(subset=['salary'])
    # in order to look at maximum salary increasing over time, we can use cummax() - cummulative max function
    
    sal = df['salary']
    return sal.cummax().shift()
```

```python
def salary_employees(df):
    """This function returns the salary hierarchy amongst employees"""
    
    # we will drop the missing values in salary column
    
    df = df.dropna(subset=['salary'])
    # in order to look at maximum salary increasing over time, we can use cummax() - cummulative max function
    
    sal = df['salary']
    
    # we will create a filter that tests, if the current salary is greater than the cummulative max
    
    filter_salary = sal > sal.cummax().shift().fillna(0)
    return df[filter_salary]
```

### For each department, find the average % of employees that have word "Senior" in their title, ignore the case, return the series with department as index

```python
def title_senior(df):
    """This functions check for a title in the dept, calculates the average"""
    
    return df['title'].str.contains('senior',case=False).groupby(df['dept']).mean().round(3) * 100
```

```python
title_senior(emp)
```

### Find the number of employees hired on 29th February

```python
def employees_hired(df):
    """This function returns the number of employees hired on 29th February"""
    
    filter_month = df['hire_date'].dt.month == 2 #feb
    
    filter_day = df['hire_date'].dt.day == 29 #29th day in Feb
    
    return (filter_month & filter_day).sum()
```

```python
employees_hired(emp)
```

### Find the percentage of employees gender per department

```python
def employee_percentage(df):
    """This funcion returns the percentage of gender as per department"""
    
    #crosstab is used to detect frequency, count elements etc
    
    return pd.crosstab(index=df['dept'],
                      columns = df['gender'],
                      normalize= 'index').round(2) * 100

# we are normalizing it through the index, hence normalize = 'index'
```

```python
employee_percentage(emp)
```
